```python
import pandas as pd
import numpy as np

import matplotlib as mpl
import matplotlib.pyplot as plt

import seaborn as sns 
%matplotlib inline
import statsmodels.formula.api as smf
from statsmodels.api import qqplot

plt.rcParams['font.family'] = 'NanumGothic'
mpl.rcParams['axes.unicode_minus'] = False

df = pd.read_csv("./SCALE_preprossed.csv", encoding='euc-kr')
df.head()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='steel_kind', hue='scale', data=df, palette='Set2')
plt.title('steel_kind별 양품/불량 수')
plt.xlabel('steel_kind')
plt.ylabel('Count')
plt.legend(title='Label', loc='upper right')
plt.show()
```

```python
grouped_data = df.groupby(['pt_length', 'scale']).size().unstack(fill_value=0)
# display(grouped_data)
good_y = [sublist[1] for sublist in grouped_data.values]
bad_y = [sublist[0] for sublist in grouped_data.values]
plt.figure(figsize=(8, 6))
plt.plot(grouped_data.index, good_y, label="양품")
plt.plot(grouped_data.index, bad_y, label="불량")
plt.xlabel("length")
plt.ylabel("Count")
plt.legend()
plt.show()
# grouped_data.values
```
